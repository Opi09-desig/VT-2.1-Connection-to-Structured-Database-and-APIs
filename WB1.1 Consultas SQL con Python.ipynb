## 2. Agregaciones y Agrupamientos de Datos.

1. ¿Cuantas canciones, albumes y artistas se tienen registrados en la base de datos?

```python
conteo_general = {
    "Total Canciones": pd.read_sql_query("SELECT COUNT(*) FROM songs;", connection).iloc[0,0],
    "Total Álbumes": pd.read_sql_query("SELECT COUNT(*) FROM albums;", connection).iloc[0,0],
    "Total Artistas": pd.read_sql_query("SELECT COUNT(*) FROM artists;", connection).iloc[0,0]
}
```

```python
df_resumen = pd.DataFrame(list(conteo_general.items()), columns=['Categoría', 'Cantidad'])
df_resumen
```

2. Agrupa la tabla albums por artista para identificar qué cuales son los 5 artistas que tienen más albumes.

```python
query_top_artists = """
SELECT artists.name, COUNT(albums._id) AS total_albums
FROM artists
JOIN albums ON artists._id = albums.artist
GROUP BY artists.name
ORDER BY total_albums DESC
LIMIT 5;
"""

df_top_artists = pd.read_sql_query(query_top_artists, connection)
df_top_artists
```

3. Presenta el nombre de los artistas identificados en el punto anterior.

```python
df_nombres_artistas = df_top_artists[['name']]
df_nombres_artistas
```

4. ¿Cuál es el album con más canciones? Investiga sobre este album en la red.

```python
query_max_songs = """
SELECT albums.name, COUNT(songs._id) AS numero_de_canciones
FROM songs
JOIN albums ON songs.album = albums._id
GROUP BY albums.name
ORDER BY numero_de_canciones DESC
LIMIT 1;
"""
```

```python
df_album_ganador = pd.read_sql_query(query_max_songs, connection)
df_album_ganador
```

## 3. Combinación de Tablas.

1. Muestra todas el nombre canciones del album "Seventh Son Of A Seventh Son" que tiene por '_id = 420'. Presenta la lista en el *track order* (`track`).

```python
query_seventh_son = """
SELECT songs.track, songs.title
FROM songs
JOIN albums ON songs.album = albums._id
WHERE albums._id = 420
ORDER BY songs.track ASC;
"""
df_seventh_son = pd.read_sql_query(query_seventh_son, connection)
df_seventh_son
```

2. Obtén todas las canciones de Aeroesmith en orden alfabético. Solo incluye el título en la salida ¿Cuántas son?

```python
query_aerosmith = """
SELECT songs.title
FROM songs
JOIN albums ON songs.album = albums._id
JOIN artists ON albums.artist = artists._id
WHERE artists.name = 'Aerosmith'
ORDER BY songs.title ASC;
"""
```

```python
df_aerosmith = pd.read_sql_query(query_aerosmith, connection)

#conteo total
total_canciones = len(df_aerosmith)
print("Aeroesmith")
print(f"Total de canciones: {total_canciones}")
df_aerosmith
```
